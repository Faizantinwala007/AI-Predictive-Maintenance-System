# AI-Based Predictive Maintenance System using Machine Learning



🔷 2. Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

🔷 3. Dataset

```python
data = {
    'temperature': [30, 80, 45, 90, 50, 85, 40],
    'vibration': [2, 8, 3, 9, 4, 7, 3],
    'failure': [0, 1, 0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)
```

🔷 4. Train Model

```python
X = df[['temperature', 'vibration']]
y = df['failure']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

🔷 5. Prediction

```python
pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

🔷 6. Visualization

```python
import matplotlib.pyplot as plt

plt.scatter(df['temperature'], df['vibration'], c=df['failure'])

plt.xlabel("Temperature")
plt.ylabel("Vibration")
plt.title("Predictive Maintenance Data Visualization")

plt.show()
```
